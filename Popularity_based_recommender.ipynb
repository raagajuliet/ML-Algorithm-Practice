```python
places = pd.DataFrame(geodata[['placeID', 'name']])
places.head()
```

```python

rating = pd.DataFrame(frame.groupby('placeID')['rating'].mean())
rating.head()
```

```python

rating['rating_count'] = pd.DataFrame(frame.groupby('placeID')['rating'].count())
rating.head()
```

```python

rating.describe()
rating.reset_index(inplace = True)
rating.head()
```

```python
rating.sort_values(by='rating_count',ascending = False).head()
# the popular restaurant is identified
```

```python
merged = rating.merge(places, on = 'placeID')
merged.head()
```

```python
cuisine_merged = merged.merge(cuisine, on = 'placeID')
cuisine_merged.head()
```

```python
cuisine_merged.sort_values(by='rating_count', ascending=False).head()
```

```python
# User place review matrix is created 
places_crosstab = pd.pivot_table(data=frame, index='userID', columns='placeID', values='rating')
places_crosstab.head()
```

```python
highest = places_crosstab[135085]
highest
```

```python
highest[highest >= 0]
```

```python
# the popular place rating is correlated with existing ratings
similar_to_highest = places_crosstab.corrwith(highest)
corr_highest = pd.DataFrame(similar_to_highest, columns=['PearsonR'])
corr_highest.dropna(inplace=True)
corr_highest.head()
```

```python
highest_corr_summary = corr_highest.merge(rating, on='placeID')
highest_corr_summary
```

```python
highest_corr_summary[highest_corr_summary['rating_count']>=10].sort_values('PearsonR',ascending=False).head()
```

```python
highest_corr_summary.sort_values('PearsonR', ascending=False).head()
```

```python
summary = highest_corr_summary.merge(cuisine, on='placeID')
summary.sort_values('PearsonR', ascending=False).head()
```
